```python
import pandas as pd
data = pd.read_csv('net_gas.csv')
data.head(2)
```

```python
date = pd.to_datetime(data['Dates'])
```

```python
def prediction_model():  
    try:
        #Injections Area
        
        injection = pd.to_datetime(input('Injection Dates: '))
        Year = injection.year
        Month = injection.month
        price = data[(data['Year']==Year) & (data['Month']==Month)]['Prices'].iloc[0]
        print(f'Injection Rate: ${price}')

    except ValueError:
            print('Enter Valid Injection Date in YYYY-MM-DD')

    try:
        #Injection Limit
        injection_limit = 1000
        inj_amt = int(input('Amount you want to buy: '))
        while inj_amt>injection_limit:
            print(f'Injection Limite per Month: {injection_limit}')
            inj_amt = int(input('Amount you want to buy: '))
        print(f'Total Injection Purchased Value: ${inj_amt*price}')
        
    except ValueError:
        print('Enter valid amount in number')
        
    try:
        #Withdraw Rate
        withdrawl = pd.to_datetime(input('Withdrawl Date: '))
        Year1 = withdrawl.year
        Month1 = withdrawl.month
        price1 = data[(data['Year']==Year1) & (data['Month']==Month1)]['Prices'].iloc[0]
        print(f'Withdrawl Rate: ${price1}')

    except ValueError:
        print('Enter Valid Withdrawl Date in YYYY-MM-DD')

    #Storage
    storage_time = ((withdrawl-injection).days)/30
    rate = 0.1

    #Storages Limits
    print(f'Stock Available to sell: {inj_amt}')
    storage_cost = rate*inj_amt*storage_time
    print(f'Storage Cost:{storage_cost}')

    try:
        draw_amt = int(input('Amount you want to sell: '))
        while draw_amt>inj_amt:
            print(f"You only have {inj_amt} stock available to sell.")
            draw_amt = int(input('Amount you want to sell: '))
        print(f'Your Withdrawl Amount: ${(draw_amt*price1)-(draw_amt*price)-(rate*draw_amt*storage_time)}')

    except ValueError:
        print('Enter valid amount in number')    
```

```python
user_inp = int(input('How many predictions do you want?: '))

for i in range(user_inp):
    prediction_model()
```
