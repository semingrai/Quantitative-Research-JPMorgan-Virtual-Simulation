```python
import pandas as pd
data = pd.read_csv("Task 3 and 4_Loan_Data (1).csv")
fico = data.fico_score
```

```python
fico_score = pd.Series(fico, name='Fico_score')
```

```python
fico_score
```

```python
five = []
four = []
three = []
two = []
one = []
for i in fico_score:
    if i >= 700:
      five.append(i)
    elif i >= 650 and i < 700:
       four.append(i)
    elif i >= 600 and i < 650:
       three.append(i)
    elif i >= 500 and i <600:
       two.append(i)
    elif i < 500:
       one.append(i)
```

```python
print(f'1 star: {sum(five)}')
print(f'2 star: {sum(four)}')
print(f'3 star: {sum(three)}')
print(f'4 star: {sum(two)}')
print(f'5 star: {sum(one)}')
```

```python
'''Fico Score:
1 = excellent
2 = very good
3 = decent
4 = bad
5 = very bad
 

FICO scores were categorized into five buckets, with higher buckets representing worse credit risk, to help summarize borrower credit profiles for modeling purposes.
'''

		

```
